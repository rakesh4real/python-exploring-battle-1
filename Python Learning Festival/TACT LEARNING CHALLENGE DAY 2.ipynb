**SLICING**

```python
# Python program

# Creating a List 
List = ['A','B','C','D','E','F', 
		'G','H','I','J','K','L','M'] 
print("Intial List: ") 
print(List) 

# Print elements of a range 
# using Slice operation 
Sliced_List = List[3:8] 
print("\nSlicing elements in a range 3-8: ") 
print(Sliced_List) 

# Print elements from beginning 
# to a pre-defined point using Slice 
Sliced_List = List[:-6] 
print("\nElements sliced till 6th element from last: ") 
print(Sliced_List) 

# Print elements from a 
# pre-defined point to end 
Sliced_List = List[5:] 
print("\nElements sliced from 5th "
	"element till the end: ") 
print(Sliced_List) 

# Printing elements from 
# beginning till end 
Sliced_List = List[:] 
print("\nPrinting all elements using slice operation: ") 
print(Sliced_List) 

# Printing elements in reverse 
# using Slice operation 
Sliced_List = List[::-1] 
print("\nPrinting List in reverse: ") 
print(Sliced_List) 
```

# III. IP Address and DNS 

**A. IP Address**

The python module **ipaddress** is used extensively to validate and categorize IP address to IPV4 and IPV6 type.
It can also be used to do comparison of the IP address values as well as IP address arithmetic for manipulating the ip addresses.

1. Validate the IPV4 Address
2. Validate the IPV6 Address
3. Check the type of IP Address
4. Comparison of IP Addresses
5. IP Addresses Arithmetic

**Validate IP Address**

```python
# import library
import ipaddress as ipaddress

# The ip_address function validates the IPV4 address. 
# If the range of values is beyond 0 to 255, then it throws an error.

# valid IPV4
print(ipaddress.ip_address(u'192.168.0.255'))

# invalid IPV4
print(ipaddress.ip_address(u'192.168.0.256'))
```

```python
# The ip_address function validates the IPV6 address.
# If the range of values is beyond 0 to ffff, then it throws an error.

# valid IPV6 address
print (ipaddress.ip_address(u'FFFF:9999:2:FDE:257:0:2FAE:112D'))

# invalid IPV6 address
print (ipaddress.ip_address(u'FFFF:10000:2:FDE:257:0:2FAE:112D'))
```

**Check the type of IP Address**

```python
# Checking IPV4 or IPV6
print(type(ipaddress.ip_address(u'192.168.0.255')))
print(type(ipaddress.ip_address(u'2001:db8::')))
```

**Comparison of IP Addresses**

```python
# Greater than
print (ipaddress.IPv4Address(u'192.168.0.2') > ipaddress.IPv4Address(u'192.168.0.1'))

# Equality
print (ipaddress.IPv4Address(u'192.168.0.2') == ipaddress.IPv4Address(u'192.168.0.1'))

# Inequality
print (ipaddress.IPv4Address(u'192.168.0.2') != ipaddress.IPv4Address(u'192.168.0.1'))
```

**IP Addresses Arithmetic**

```python
# Addition
print (ipaddress.IPv4Address(u'192.168.0.2')+1)

# Subtraction
print (ipaddress.IPv4Address(u'192.168.0.253')-3)

# Increases the previous octet by value 1.
print (ipaddress.IPv4Address(u'192.168.10.253')+3)

# Throws Value error
print (ipaddress.IPv4Address(u'255.255.255.255')+1)
```

**B. DNS Look-up**

The translation of domain names to IP address is managed by the python module **dnspython**.
This module also provides methods to find out CNAME and MX records.

1. Finding 'A' Record
> The mapping between IP address and domain name is also known as 'A' record.

2. Finding CNAME Value
> Canonical Name Record is a type of record in the Domain Name System (DNS) used to map a domain name as an alias for another domain

3. Finding MX Record
> Mail Exchanger record specifies a mail server responsible for accepting email messages on behalf of a recipient's domain

```python
import dnspython as dns
import dns.resolver
```

```python
# Program for finding 'A' Record

# import libraries
import dnspython as dns
import dns.resolver

result = dns.resolver.query('talentaccurate.com', 'A')
for ipval in result:
    print('IP', ipval.to_text())
```

```python
# Program to find CNAME values

# import libraries
import dnspython as dns
import dns.resolver

result = dns.resolver.query('mail.google.com', 'CNAME')
for cnameval in result:
    print ('cname target address:', cnameval.target)
```

```python
# Program to find MX Records

result = dns.resolver.query('hotmail.com', 'MX')
for exdata in result:
    print('MX Record:', exdata.exchange.to_text())
```

 # V. Simple CRUD in sqlite 




Python has built-in support for SQLite in the form of the sqlite3 module. This module contains functions for performing persistent CRUD operations on SQLite database.

As per the prescribed standards, the first step in the process is to obtain the connection. In order to establish a connection with a SQLite database, 

> **sqlite3** module needs to be imported and the connect() function needs to be executed.

**Note** : The connect() function returns a connection object referring to the existing database or a new database if it doesn't exist.

A **cursor** is a Python object that enables you to work with the database.	It acts as a handle for a given SQL query; it allows the retrieval of one or more rows of the result. 
Hence, a cursor object is obtained from the connection to execute SQL queries using the following statement:
> cur = db.cursor()

### Importing relevant libraries

```python
import sqlite3
import pandas as pd
```

### Obtain connection

```python
db=sqlite3.connect('test.db')
```

### Initialize a cursor

```python
cur=db.cursor()
```

### Create new table

```python
# Program to create a new table

# import libraries
import sqlite3

# connection
db=sqlite3.connect('test.db')

try:        
    # cursor
    cur =db.cursor()
    
    # query
    cur.execute('''CREATE TABLE students0 (
    StudentID INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT (20) NOT NULL,
    age INTEGER,
    marks REAL);''')
    
    # prompt sucess
    print ('table created successfully')
    
except:
    # prompt failure
    print ('error in operation')
    
    # to starting point
    db.rollback()
    
db.close()
```

### Insert into table

1. **execute( )** : Single value insertion


2. **executemany( )** : Multiple value insertion

```python
# program to insert 
# SINGLE VALUE into table

# connection
db=sqlite3.connect('test.db')

# query
qry="insert into students0 (name, age, marks) values(?,?,?);"

try:
    # cursor
    cur=db.cursor()
    
    # single value insertion
    cur.execute(qry, ('Vijaya', 16,75))
    
    # commit changes
    db.commit()
    
    # prompt sucess
    print ("one record added successfully")
    
except:
    # prompt failure
    print("error in operation")
    
    # to starting point
    db.rollback()
    
db.close()
```

```python
# Program to insert
# MULTIPLE VALUES into table

# connection
db=sqlite3.connect('test.db')

# Query and values to be added
qry="insert into students0 (name, age, marks) values(?,?,?);"
students=[('Amar', 18, 70), ('Deepak', 25, 87)]

try:
    # cursor
    cur=db.cursor()
    
    # multiple value insertion
    cur.executemany(qry, students)
    
    #commit changes
    db.commit()
    
    # prompt success
    print ("records added successfully")
    
except:
    # prompt failure
    print ("error in operation")
    
    # to starting point
    db.rollback()
    
db.close()
```

### Retrieve Records from a table



1. **fetchone( )**: Fetches the next available record from the result set. It is a tuple consisting of values of each column of the fetched record.


2. **fetchall()**: Fetches all remaining records in the form of a list of tuples. Each tuple corresponds to one record and contains values of each column in the table.

```python
# Python program to Retrieve records using
# fetchone()

# connection
db=sqlite3.connect('test.db')

# query
sql="SELECT * from students0;"

# cursor
cur=db.cursor()

# execute query
cur.execute(sql)

#loop to fetch
while True:
    record=cur.fetchone()
    if record==None:
        break
    print (record)
    
db.close()
```

```python
# Python program to Retrieve records using
# fetchone()

# connection
db=sqlite3.connect('test.db')

# query
sql="SELECT * from students0;"

# cursor
cur=db.cursor()

# execute query
cur.execute(sql)

#loop to fetch all records
while True:
    students=cur.fetchall()
    for rec in students:
        print(rec)
        
db.close()
```

### Update a record

```python
# Python program to
# Update a record

# connection
db=sqlite3.connect('test.db')

# update query
qry="update students0 set age=? where name=?;"

try:
    # initialize cursor
    cur=db.cursor()
    
    # execute query
    cur.execute(qry, (19,'Deepak'))
    
    # commit changes
    db.commit()
    
    # prompt success
    print("record updated successfully")
    
except:
    # Prompt failure
    print("error in operation")
    
    # to starting point
    db.rollback()
    
db.close()
```

### Delete Record

```python
# Python program to 
# Delete a record

# Connection
db=sqlite3.connect('test.db')

# Delete query
qry="DELETE from students0 where name=?;"

try:
    # Execute cursor
    cur=db.cursor()
    
    # delete
    cur.execute(qry, ('Bill',))
    
    #commit changes
    db.commit()
    
    #prompt succcess
    print("record deleted successfully")
    
except:
    # prompt failure
    print("error in operation")
    
    # to starting point
    db.rollback()
    
db.close()
```

# VI. Lambda Functions

The power of lambda is better shown when you use them as an anonymous function inside another function

```python
# Simle function
def times2(var):
    return var*2

# calling simple function
times2(2)
```

```python
# Lambda expression
lambda var: var*2
```

```python
# MAPS AND FILTERS

seq = [1,2,3,4,5]

print(map(times2,seq))

print(list(map(times2,seq)))

print(list(map(lambda var: var*2,seq)))

print(filter(lambda item: item%2 == 0,seq))

print(list(filter(lambda item: item%2 == 0,seq)))
```

```python
# create lambda function
def myfunc(n):
    # lambda expression
    return lambda a : a * n

mydoubler = myfunc(2)
mytripler = myfunc(3)

print(mydoubler(11)) 
print(mytripler(11))
```
